# printing the goal difference of all teams

```python
sports['BasketScored'] = sports['BasketScored'].astype(int)
sports['BasketGiven'] = sports['BasketGiven'].astype(int)
```

```python
def Basket_different():   
     sports['Basket_different'] = sports['BasketScored']-sports['BasketGiven']
     return sports[['Team','Basket_different']]
```

```python
Basket = Basket_different()
```

```python
Basket.sort_values(by = 'Basket_different',ascending=False)
```

# Collecting wining percentage of the data set

```python
sports['PlayedGames'] = sports['PlayedGames'].astype(int)
sports['WonGames'] = sports['WonGames'].astype(int)
```

```python
sports['Win'] = (sports['WonGames']/sports['PlayedGames']) *100
```

```python
sports['Win'].fillna(0,inplace = True)
```

```python
sports[['Team','Win']]
```
